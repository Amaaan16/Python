```python
#What is the probability of winning a match if the toss was won?
```

```python
total_matches = len(matches)
toss_won_matches = len(matches[matches['toss_winner'] == matches['winner']])
probability_of_winning_given_toss_won = toss_won_matches / total_matches
print("Probability of winning a match after winning toss:",probability_of_winning_given_toss_won*100,"%")
```

```python
matches['toss_win_and_match_win'] = (matches['toss_winner'] == matches['winner']).astype(int)
probability_df = matches.groupby('season')['toss_win_and_match_win'].mean().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='season', y='toss_win_and_match_win', data=probability_df, marker='o', color='blue')
plt.xlabel('Season')
plt.ylabel('Probability of Winning Given Toss Won')
plt.title('Probability of Winning a Match Given Toss Won Over Seasons')
plt.xticks(rotation=45, ha='right')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='season', y='toss_win_and_match_win', data=probability_df, color='skyblue')
plt.xlabel('Season')
plt.ylabel('Probability of Winning Given Toss Won')
plt.title('Probability of Winning a Match Given Toss Won Over Seasons')
plt.xticks(rotation=45, ha='right')
```

```python
#The total runs by fours hit and the total number of fours hit by each team
```

```python
fours_data = deliveries.groupby('batting_team')['total_runs'].agg(['sum', lambda x: (x == 4).sum()]).reset_index()
fours_data.columns = ['team', 'total_runs_by_fours', 'total_fours']
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='team', y='total_runs_by_fours', data=fours_data, label='Total Runs by Fours')
plt.xlabel('Team')
plt.ylabel('Count')
plt.title('Total Runs by Fours Hit by Each Team')
plt.xticks(rotation=45, ha='right')
plt.legend()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='team', y='total_fours', data=fours_data, label='Total Fours')
plt.xlabel('Team')
plt.ylabel('Count')
plt.title('Total Number of Fours Hit by Each Team')
plt.xticks(rotation=45, ha='right')
plt.legend()
```

```python
#The fours hit by players
```

```python
merged_data = pd.merge(matches, deliveries, left_on='id', right_on='match_id')
merged_data = merged_data[merged_data['result'] == 'normal']
fours_by_player = merged_data[merged_data['batsman_runs'] == 4].groupby('batsman')['batsman_runs'].count().reset_index()
fours_by_player.columns = ['player', 'total_fours']
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x='player', y='total_fours', data=fours_by_player.sort_values(by='total_fours', ascending=False).head(10), palette='viridis')
plt.xlabel('Player')
plt.ylabel('Number of Fours')
plt.title('Top 10 Players by Number of Fours Hit')
plt.xticks(rotation=45, ha='right')
```

```python
#The number of fours hit in each season
```

```python
fours_by_season = merged_data[merged_data['batsman_runs'] == 4].groupby('season')['batsman_runs'].count().reset_index()
fours_by_season.columns = ['season', 'total_fours']
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='season', y='total_fours', data=fours_by_season, palette='viridis')
plt.xlabel('Season')
plt.ylabel('Number of Fours')
plt.title('Number of Fours Hit in Each IPL Season')
plt.xticks(rotation=45, ha='right')
```

```python
#The total runs by the sixes hit and the number of sixes hit by each team
```

```python
sixes_data = merged_data[merged_data['batsman_runs'] == 6].groupby('batting_team')['batsman_runs'].agg(['sum', 'count']).reset_index()
sixes_data.columns = ['team', 'total_runs_by_sixes', 'total_sixes']
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='team', y='total_runs_by_sixes', data=sixes_data, label='Total Runs by Sixes')
plt.xlabel('Team')
plt.ylabel('Count')
plt.title('Total Runs by Sixes Hit by Each Team')
plt.xticks(rotation=45, ha='right')
plt.legend()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='team', y='total_sixes', data=sixes_data, label='Total Sixes')
plt.xlabel('Team')
plt.ylabel('Count')
plt.title('Total Number of Sixes Hit by Each Team')
plt.xticks(rotation=45, ha='right')
plt.legend()
```

```python
#The number of sixes hit in each season
```

```python
sixes_by_season = merged_data[merged_data['batsman_runs'] == 6].groupby('season')['batsman_runs'].count().reset_index()
sixes_by_season.columns = ['season', 'total_sixes']
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='season', y='total_sixes', data=sixes_by_season, palette='viridis')
plt.xlabel('Season')
plt.ylabel('Number of Sixes')
plt.title('Number of Sixes Hit in Each IPL Season')
plt.xticks(rotation=45, ha='right')
```

```python
#The top 10 leading run scorers in IPL
```

```python
total_runs_by_batsman = merged_data.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
top_10_scorers = total_runs_by_batsman.sort_values(by='batsman_runs', ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='batsman', y='batsman_runs', data=top_10_scorers, palette='viridis')
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.title('Top 10 Leading Run Scorers in IPL')
plt.xticks(rotation=45, ha='right')
```
